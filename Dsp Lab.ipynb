# Lab 5: Convolution and Correlation

```python
#Convolution of Two signal
import numpy as np
signal1=np.array([2,1,0,1])
signal2=np.array([1,2,3,1])
convolution_result=np.convolve(signal1,signal2,mode='full')
print('Convolution Result: ', convolution_result)
```

```python
#Auto correlation
signal1=np.array([1,2,3,4,5,1])
correlation_result=np.correlate(signal1,signal1,mode='full')
print('Correlation Result ',correlation_result)
```

```python
#Cross Correlation
signal1=np.array([2,1,0,1])
signal2=np.array([1,2,3,1])
correlation_result1=np.correlate(signal1,signal2,mode='full')
print('Correlation Result ',correlation_result1)
```

# Lab3: Analog Signal ,Digital Singal, Unit step signal,Ramp Signal and so on

```python
#sine graph
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,5*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,color='red')
plt.show()
```

```python
#cosine graph

x=np.arange(0,5*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y,color='blue')
plt.show()
```

```python
#DT signal
#x(n)=sin(2pifn/Fs)
L=50
f=10
Fs=500
x=np.zeros(L)
for n in range(L):
    x[n]=np.sin(2*np.pi*f*n/Fs)  #t=nT=n/Fs
plt.stem(x)
```

```python
#impulse signal del(n) is 1 when n=0 otherwise always zero
d=np.zeros(L)
d[L//2]=1
x=np.arange(-L//2,L//2)
plt.stem(x,d)
```

```python
#dell(n-10)
d=np.zeros(L)
d[L//2+10]=1
x=np.arange(-L//2,L//2)
plt.stem(x,d)

```

```python
#del(-n-10)
d=np.zeros(L)
d[L//2-10]=1
x=np.arange(-L//2,L//2)
plt.stem(x,d)
```

```python
#Ramp signal
x=np.zeros(L)
for i in range(L):
    x[i]=i
plt.stem(x)
```

```python
#negative ramp
d=np.zeros(L)
x=np.zeros(L)
for i in range(L):
    x[i]=-i
for i in range(L):
    d[i]=i
plt.stem(x,d)


```

```python
#unit signal
x=np.zeros(L);
for i in range(L):
    x[i]=1;
plt.stem(x)
```

```python
#unit signal x(-n)

x=np.zeros(L)
d=np.zeros(L)
for i in range(L):
    x[i]=-i
for i in range(L):
    d[i]=1
plt.stem(x,d)
```

```python
#shifting property x(n-10) range will be from 10 to L+10

x=np.zeros(L+12)
d=np.zeros(L+12)
for i in range(10,L+10+1):
    x[i]=i   
for i in range(10,L+10+1):
    d[i]=1
plt.stem(x,d)
```

```python
#exponential signal
# x(n)=a^n
a=0.95
x=np.zeros(L)
for n in range(L):
    x[n]=a**n
plt.stem(x)
```

```python
a=-0.95
x=np.zeros(L)
for i in range(L):
    x[i]=a**i
plt.stem(x)
```

```python
a=1.2
x=np.zeros(L)
for i in range(L):
    x[i]=a**i
plt.stem(x)
```

# Lab 4: Noise ,Moving average filtering,first order difference( Noise Filtering)

```python

L=300
f=10
Fs=300
x=np.arange(L)
s = np.sin(2 * np.pi *x *f/Fs)
plt.plot(x,s)
plt.grid('true')
plt.xlabel('Sample')
```

```python
import random
noise=np.zeros(L)
for i in range(L):
    noise[i]=random.random()
plt.plot(noise)
```

```python
ns=s+noise
plt.plot(ns)
```

```python
#moving average filtering
x=np.zeros(L)
for i in range(1,L-1):
    x[i]=(ns[i-1]+ns[i]+ns[i+1])/3
    
plt.plot(x)
```

```python
#accumulator
y=np.zeros(L)
for k in range(L):
    y[k]=np.sum(x[0:k])
plt.plot(y)
```

```python
z=np.zeros(L)
for i in range(1,L):
    z[i]=y[i]-y[i-1]
plt.plot(z)
plt.grid('true')
plt.title('difference')
plt.xlabel('Sample')
plt.ylabel('O/p')
plt.show()
```
